```python
# imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Initial params

```python
# Make empty image
img = np.zeros((800,800,3),dtype=np.uint8)
rows,cols,_=img.shape
# By increasing this number, the final shape will be made of more lines.
numofsteps=22
```

## The logic of shape creation is implemented in this part.

```python
asteps = np.empty(2, object)
asteps[0] = np.linspace(0, cols - 1, numofsteps + 1, dtype=np.int16)
asteps[1] = np.linspace(0, rows - 1, numofsteps + 1, dtype=np.int16)
regionColor=np.array(((255,0,255),(0,255,0),(0,255,255),(255,0,255)),dtype=np.uint8)
midpoints = [(cols // 2, 0), (cols - 1, rows // 2), (cols // 2,rows - 1), (0, rows // 2)]

for i in range(0, 4):
    pt1 = midpoints[i]
    B,G,R=map(int,regionColor[i])
    for j in range(0, numofsteps+1):
        if i in (0,2):
            pt2=(asteps[0][j],rows//2)
        elif i in (1,3):
            pt2 = (cols // 2,asteps[1][j])
        cv2.line(img, pt1, pt2, (B,G,R), thickness=1)
```

## Initialize matplotlib

```python
figure=plt.figure()
```

## Show result

```python
figure.add_subplot(1,1,1)
plt.imshow(img)
plt.axis('off')
plt.title('Result')
```
