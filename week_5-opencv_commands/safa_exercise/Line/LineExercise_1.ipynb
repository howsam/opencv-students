```python
# imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Initial params

```python
# Make empty image
img = np.zeros((800,800,3),dtype=np.uint8)
rows,cols,_=img.shape
# By increasing this number, the final shape will be made of more lines.
numofsteps=22
```

## The logic of shape creation is implemented in this part.

```python
# Making a square with four sides
aEdges=np.empty(4,dtype=object)
# Based on the numofsteps, we determine the points on the sides of the square.
aEdges[0]=np.linspace(0,cols-1,numofsteps+1,dtype=np.int16)
aEdges[1]=np.linspace(0,rows-1,numofsteps+1,dtype=np.int16)
aEdges[2]=np.linspace(cols-1,0,numofsteps+1,dtype=np.int16)
aEdges[3]=np.linspace(rows-1,0,numofsteps+1,dtype=np.int16)
# Specifying colors
regionColor=np.array(((255,255,255),(0,255,0),(255,0,255),(0,255,255)),dtype=np.uint8)


for i in range(0,4):
    B,G,R=map(int,regionColor[i])
    for j in range(0,12):
        if i == 0:
            pt1=(aEdges[i][j],0)
            pt2 = (cols-1, aEdges[i+1][j+1])
        if i == 1:
            pt1 = (cols-1, aEdges[i][j])
            pt2 = (aEdges[i + 1][j + 1],rows-1 )
        if i == 2:
            pt1 = (aEdges[i][j],rows-1)
            pt2 = (0,aEdges[i + 1][j + 1] )
        if i == 3:
            pt1 = (0,aEdges[i][j])
            pt2 = (aEdges[0][j + 1],0)
        cv2.line(img, pt1, pt2, (B,G,R), thickness=1)
```

## Initialize matplotlib

```python
figure=plt.figure()
```

## Show result

```python
figure.add_subplot(1,1,1)
plt.imshow(img)
plt.axis('off')
plt.title('Result')
```
