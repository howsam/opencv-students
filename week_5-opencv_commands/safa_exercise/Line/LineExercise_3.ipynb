```python
# imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Initial params

```python
img = np.zeros((800, 800, 3), dtype=np.uint8)
rows, cols, _ = img.shape
numofsteps = 22
regioncolor = np.array([(255,0,0), (0,255,0), (0,0,255), (128,0,128)],dtype=np.uint8)
```

## The logic of shape creation is implemented in this part.

```python
regions = np.empty((4, 2), object)
# Region 1
regions[0, 0] = np.linspace(0, rows // 2, numofsteps, dtype=np.int16)
regions[0, 1] = np.linspace(cols // 2, cols - 1, numofsteps, dtype=np.int16)

# Region 2
regions[1, 0] = np.linspace(rows - 1, rows // 2, numofsteps, dtype=np.int16)
regions[1, 1] = np.linspace(cols // 2, cols - 1, numofsteps, dtype=np.int16)

# Region 3
regions[2, 0] = np.linspace(rows // 2, rows - 1, numofsteps, dtype=np.int16)
regions[2, 1] = np.linspace(0, cols // 2, numofsteps, dtype=np.int16)

# Region 4
regions[3, 0] = np.linspace(rows // 2, 0, numofsteps, dtype=np.int16)
regions[3, 1] = np.linspace(0, cols // 2, numofsteps, dtype=np.int16)


for region in range(0, 4):
    for i in range(0, numofsteps):
        pt1 = (cols // 2, regions[region][0][i])
        pt2 = (regions[region][1][i], rows // 2)
        B, G, R = map(int, regioncolor[region])  # Convert to integers
        cv2.line(img, pt1, pt2,color=(B,G,R), thickness=1)
```

## Show result

```python
figure=plt.figure()
figure.add_subplot(1,1,1)
plt.imshow(img)
plt.axis('off')
plt.title('Result')
```
