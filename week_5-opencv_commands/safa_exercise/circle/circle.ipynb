## Imports

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## Body of excercise

```python
img=np.ones((800,800),dtype=np.float32)
rows,cols=img.shape

for i in range(1,16):
    cv2.circle(img,(cols//2,rows//2),i*20,0,thickness=20-i)
```

## Show result

```python
figure=plt.figure()
figure.add_subplot(1,1,1)
plt.imshow(img)
plt.axis('off')
plt.title('Result')
```
